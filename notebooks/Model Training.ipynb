```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('../data/processed/Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
## drop day,month and year
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Independent and Dependent Features
X = df.drop('FWI',axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,8))
corr = X_train.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
def correlation(dataset, threshold) : ## function to find correlated features
    col_corr = set() 
    corr_matrix = dataset.corr() 
    for i in range(len(corr_matrix.columns)): 
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: ## check if correlation is greater than threshold
                colname = corr_matrix.columns[i] 
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features = correlation(X_train, 0.8)
corr_features
```

```python
## drop features when correlation is more than 0.8
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler ## Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) ## fit and transform on training data
X_test_scaled = scaler.transform(X_test) ## transform on testing data
```

```python
X_train_scaled
```

## Box Plots To understand Effect Of Standard Scaler

```python
## Visualize the effect of scaling
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
lr = LinearRegression()
lr.fit(X_train_scaled,y_train) ## model training
y_pred = lr.predict(X_test_scaled) ## model prediction
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R2 Score: {r2}")
plt.scatter(y_test,y_pred)
plt.show()
```

## Lasso Regression Model

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, r2_score 
lasso = Lasso()
lasso.fit(X_train_scaled,y_train) ## model training
y_pred = lasso.predict(X_test_scaled) ## model prediction
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R2 Score: {score}")
plt.scatter(y_test,y_pred)
plt.show()
```

## Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV 
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train) ## model training
```

## Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score 
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred = ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R2 Score: {r2}")
plt.scatter(y_test,y_pred)
plt.show()
```

## Cross Validation Ridge

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train) ## model training
y_pred = ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"R2 Score: {score}")
plt.show()
```
